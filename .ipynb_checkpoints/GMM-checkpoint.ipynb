```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

### Load Data

```python
path = 'merged_df.csv'
df = pd.read_csv(path)
```

### Transformation - Units

```python
df['Vitamin A'] = df['Vitamin A'] / 1000  # µg → mg
# df[['Vitamin A', 'Vitamin B1', 'Vitamin D']]
units_in_g =['Fat', 'Saturated Fats', 'Monounsaturated Fats', 'Polyunsaturated Fats', 'Carbohydrates', 'Sugars', 'Protein', 'Dietary Fiber', 'Water']
for col in units_in_g:
    df[col] = df[col] * 1000  # g → mg
```

```python
df1 = df.copy()
food_column = df1['food']  
df1 = df1.drop(columns=['food'])  

# Standardize the data
scaler = StandardScaler()
df1_scaled = scaler.fit_transform(df1)

# Convert back to DataFrame with original column names
df1_scaled = pd.DataFrame(df1_scaled, columns=df1.columns)

# Add back the 'food' column
df1_scaled.insert(0, 'food', food_column)

print(df1_scaled.head())
```

```python
# Define feature groups
vitamin_features = [
    "Vitamin A", "Vitamin B1", "Vitamin B11", "Vitamin B12",
    "Vitamin B2", "Vitamin B3", "Vitamin B5", "Vitamin B6",
    "Vitamin C", "Vitamin D", "Vitamin E", "Vitamin K"
]

element_features = [
    "Sodium", "Calcium", "Copper", "Iron", "Magnesium",
    "Manganese", "Phosphorus", "Potassium", "Selenium", "Zinc"
]

macronutrient_features = [
    "Fat", "Saturated Fats", "Monounsaturated Fats", "Polyunsaturated Fats",
    "Carbohydrates", "Sugars", "Protein", "Dietary Fiber",
    "Cholesterol", "Sodium", "Water"
]

# Feature sets for each GMM run
feature_sets = {
    "all_features": df1_scaled.drop(columns=["food"]),
    "without_nutrition_density": df1_scaled.drop(columns=["food", "Nutrition Density"]),
    "vitamins": df1_scaled[vitamin_features],
    "elements": df1_scaled[element_features],
    "macronutrients": df1_scaled[macronutrient_features]
}
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

def run_gmm(X, feature_set_name):
    print(f"\nRunning GMM for: {feature_set_name}")

    # Determine the optimal number of clusters using BIC
    bic_scores = []
    n_components_range = range(2, 11)

    for n in n_components_range:
        gmm = GaussianMixture(n_components=n, covariance_type="full", random_state=42)
        gmm.fit(X)
        bic_scores.append(gmm.bic(X))

    # Find the best number of clusters (lowest BIC)
    optimal_clusters = n_components_range[bic_scores.index(min(bic_scores))]
    print(f"Optimal clusters for {feature_set_name}: {optimal_clusters}")

    # Fit GMM with optimal clusters
    gmm = GaussianMixture(n_components=optimal_clusters, covariance_type="full", random_state=42)
    gmm.fit(X)
    cluster_labels = gmm.predict(X)

    # Add cluster labels to the DataFrame
    df1_scaled[f"GMM_Cluster_{feature_set_name}"] = cluster_labels

    # Save clustered data
    output_file = f"GMM_clustered_{feature_set_name}.csv"
    df1_scaled.to_csv(output_file, index=False)
    print(f"Clustered data saved as '{output_file}'")

    # PCA for visualization
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)

    # Convert to DataFrame
    X_pca_df = pd.DataFrame(X_pca, columns=["PCA1", "PCA2"])
    X_pca_df["Cluster"] = cluster_labels

    # Scatter plot of clusters
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=X_pca_df["PCA1"], y=X_pca_df["PCA2"], hue=X_pca_df["Cluster"], palette="viridis", s=100, edgecolors="k")
    plt.title(f"GMM Clustering ({feature_set_name}) - PCA Reduced")
    plt.xlabel("Principal Component 1")
    plt.ylabel("Principal Component 2")
    plt.legend(title="Cluster")
    plt.show()
```

```python
for feature_set_name, X in feature_sets.items():
    run_gmm(X, feature_set_name)
```

```python
def summarize_clusters(df, feature_set_name):
    cluster_col = f"GMM_Cluster_{feature_set_name}"  # The target cluster column

    if cluster_col not in df.columns:
        print(f"Skipping {feature_set_name} - No clustering data found.")
        return

    # Drop non-numeric columns and other GMM cluster labels
    numeric_df = df.drop(columns=["food"] + [col for col in df.columns if "GMM_Cluster" in col and col != cluster_col], errors="ignore")

    # Group data by the selected clustering and calculate mean
    cluster_summary = numeric_df.groupby(cluster_col).mean()

    # Identify the cluster with the highest value for each feature
    highest_cluster_per_feature = cluster_summary.idxmax(axis=0)


    print("\n Cluster Mean Summary:")
    print(cluster_summary.T)  

    print("\n Feature Dominance (Highest Cluster Per Feature):")
    print(highest_cluster_per_feature)
```

```python
for feature_set_name in feature_sets.keys():
    summarize_clusters(df1_scaled, feature_set_name)
```

```python
import pandas as pd

def sample_foods_by_cluster(df, feature_set_name, n_samples=5):
    cluster_col = f"GMM_Cluster_{feature_set_name}"  # Target cluster column

    if cluster_col not in df.columns:
        print(f"Skipping {feature_set_name} - No clustering data found.")
        return

    print(f"\n Sample Foods for {feature_set_name}:")
    
    # Get unique clusters
    clusters = df[cluster_col].unique()
    
    for cluster in sorted(clusters):
        cluster_foods = df[df[cluster_col] == cluster]["food"]
        
        # If the cluster has fewer than n_samples, take all
        sampled_foods = cluster_foods.sample(min(n_samples, len(cluster_foods)), random_state=42)
        
        print(f"\n Cluster {cluster}:")
        print(sampled_foods.to_list())  # Print as list for readability
```

```python
gmm_methods = ["all_features", "without_nutrition_density", "vitamins", "elements", "macronutrients"]

for method in gmm_methods:
    sample_foods_by_cluster(df1_scaled, method, n_samples=5)
```
